```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store CSV files into DataFrames

```python
# 2015 data
clean15_file = "../Data_Excel/Clean/clean_2015.csv"
data15_df = pd.read_csv(clean15_file)
data15_df.head()
```

```python
# 2016 data
clean16_file = "../Data_Excel/Clean/clean_2016.csv"
data16_df = pd.read_csv(clean16_file)
data16_df.head()
```

```python
# 2017 data
clean17_file = "../Data_Excel/Clean/clean_2017.csv"
data17_df = pd.read_csv(clean17_file)
data17_df.head()
```

```python
# 2018 data
clean18_file = "../Data_Excel/Clean/clean_2018.csv"
data18_df = pd.read_csv(clean18_file)
data18_df.head()
```

```python
# 2019 data
clean19_file = "../Data_Excel/Clean/clean_2019.csv"
data19_df = pd.read_csv(clean19_file)
data19_df.head()
```

### Add year column to dataframes

```python
data15_df['year'] = 2015
data15_df=data15_df.reindex(['year', 'country', 'happiness_rank',
                           'happiness_score', 'gdp_per_capita', 'family',
                           'life_expectancy', 'freedom', 'generosity', 'government_corr'], axis=1)
data15_df.head()


data16_df['year'] = 2016 
data16_df=data16_df.reindex(['year', 'country', 'happiness_rank',
                           'happiness_score', 'gdp_per_capita', 'family',
                           'life_expectancy', 'freedom', 'generosity', 'government_corr'], axis=1)
data16_df.head()

data17_df['year'] = 2017
data17_df=data17_df.reindex(['year', 'country', 'happiness_rank',
                           'happiness_score', 'gdp_per_capita', 'family',
                           'life_expectancy', 'freedom', 'generosity', 'government_corr'], axis=1)
data17_df.head()

data18_df['year'] = 2018
data18_df=data18_df.reindex(['year', 'country', 'happiness_rank',
                           'happiness_score', 'gdp_per_capita', 'social_support',
                           'life_expectancy', 'freedom', 'generosity', 'government_corr'], axis=1)
data18_df.head()

data19_df['year'] = 2019
data19_df=data19_df.reindex(['year', 'country', 'happiness_rank',
                           'happiness_score', 'gdp_per_capita', 'social_support',
                           'life_expectancy', 'freedom', 'generosity', 'government_corr'], axis=1)
data19_df.head()
```

### Merge data using year column

```python
# add 2015-2017 data to new dataframe
data15_17=data15_df.append([data16_df, data17_df])
# data15_17.head()

# add 2018-2019 data to new dataframe
data18_19=data18_df.append(data19_df)
# data18_19.head()

# combine files 
allData_df=data15_17.append([data18_19], sort=False)
# allData_df.head()

allData_df.reindex(['year', 'country', 'happiness_rank','happiness_score', 
                    'gdp_per_capita', 'family', 'social_support',
                    'life_expectancy', 'freedom', 'generosity', 'government_corr'], axis=1)
allData_df
```

```python
# create .csv file from dataframe
allData_df.to_csv("../Data_Excel/Clean/clean_all.csv", index=False)
```
